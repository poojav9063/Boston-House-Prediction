```python
train_data=train_data[~((train_data['crim']>=25))]
train_data=train_data[~((train_data['rm']>=8.5)|(train_data['rm']<4))]
train_data=train_data[~((train_data['lstat']>=30))]
```

```python
X_DataFrame = train_data[['lstat','rm','indus','ptratio','tax','nox','crim','age']]
X = X_DataFrame.values
X
```

```python
y_DataFrame = train_data[['medv']]
y = y_DataFrame.values
y
```

```python
LR = LinearRegression()
LR.fit(X,y)
```

```python
print("Intercept : ", LR.intercept_)
print("Slope : ", LR.coef_)
```

```python
Y_pred = LR.predict(X)
print("Predicted Values :\n",Y_pred)
```

```python
RMSE = np.sqrt(mean_squared_error(y,Y_pred))
print("Root Mean Square Error : ", RMSE)
```

```python
X_test = test_data[['lstat','rm','indus','ptratio','tax','nox','crim','age']].values
Y_test_pred = LR.predict(X_test)
for x, y in zip(test_data['ID'],Y_test_pred):
    print(x, y, sep='\t\t')
```
